Here is a list of some common anomaly detection techniques used:

- Statistical Methods:
   - Z-score: Detect anomalies based on the number of standard deviations away from the mean.
   - Modified Z-score: A variation of Z-score that is robust to outliers.
   - Percentiles: Identify anomalies based on data points falling outside a specified percentile range.
   - Grubbs' Test: Detect outliers in a univariate dataset based on extreme values.
   
- Density-Based Methods:
   - Local Outlier Factor (LOF): Measures the local density deviation of a data point with respect to its neighbors.
   - DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Clusters the data and labels points with low-density neighborhoods as outliers.
   

- Distance-Based Methods:
   - k-Nearest Neighbors (k-NN): Assign anomalies based on the distance to their k-nearest neighbors.
   - Mahalanobis Distance: Measures the distance between a data point and the center of a distribution, accounting for the covariance between variables.


- Machine Learning-Based Methods:
   - One-Class SVM (Support Vector Machine): Trains a model on normal data and identifies deviations as anomalies.
   - Autoencoders: Unsupervised neural network models that learn to reconstruct normal data and flag deviations as anomalies.
   - Isolation Forest: Builds an ensemble of isolation trees to isolate anomalies.
   

- Time Series Anomaly Detection:
   - Moving Averages: Apply a moving average to smooth out noise in the time series. Anomalies are detected based on the deviation from the moving average.
   - Seasonal Decomposition: Decompose a time series into its components (trend, seasonality, and residual). Anomalies can be detected in the residual component.
   - Exponential Smoothing: Forecasts future values and identifies anomalies based on deviations from the forecasted values.
   - ARIMA (AutoRegressive Integrated Moving Average): Fit an ARIMA model to the time series data and compare the residuals to detect anomalies.
   - LSTM Autoencoders: Uses Long Short-Term Memory networks to detect anomalies in sequential data.
   

- Multivariate Anomaly Detection:
   - Principal Component Analysis (PCA): Reduces dimensionality and identifies anomalies based on the residual errors.


통계적 방법:
- Z-score: 평균으로부터 표준편차 수를 기준으로 이상치 탐지.
- Modified Z-score: Z-score의 변형, 이상치에 강함.
- Percentiles: 지정된 백분위 범위를 벗어난 데이터 포인트로 이상치 식별.
- Grubbs' Test: 단변량 데이터셋에서 극값을 기반으로 이상치 탐지.

밀도 기반 방법:

- Local Outlier Factor (LOF): 데이터 포인트의 이웃에 대한 지역 밀도 편차 측정.
- DBSCAN: 데이터를 클러스터링하고 저밀도 이웃을 이상치로 라벨링.

거리 기반 방법:

- k-Nearest Neighbors (k-NN): k-최근접 이웃까지의 거리를 기준으로 이상치 할당.
- Mahalanobis Distance: 데이터 포인트와 분포 중심 간의 거리 측정, 변수 간 공분산 고려.

머신러닝 기반 방법:

- One-Class SVM: 정상 데이터를 학습하고 편차를 이상치로 식별.
- Autoencoders: 정상 데이터를 재구성하도록 학습하는 비지도 신경망 모델, 편차를 이상치로 플래그.
- Isolation Forest: 트리 앙상블을 구축하여 이상치 격리.

시계열 이상치 탐지:

- Moving Averages: 이동 평균을 적용하여 시계열의 잡음 제거. 이동 평균으로부터의 편차로 이상치 탐지.
- Seasonal Decomposition: 시계열을 구성 요소(추세, 계절성, 잔차)로 분해. 잔차 구성 요소에서 이상치 탐지.
- Exponential Smoothing: 미래 값을 예측하고 예측 값으로부터의 편차로 이상치 식별.
- ARIMA: 시계열 데이터에 ARIMA 모델을 맞추고 잔차를 비교하여 이상치 탐지.
- LSTM Autoencoders: 순차 데이터에서 이상치를 탐지하기 위해 Long Short-Term Memory 네트워크 사용.

다변량 이상치 탐지:

- Principal Component Analysis (PCA): 차원 축소 및 잔차 오류를 기반으로 이상치 식별.

# Z-score anomalies (Statistical Method)

- **Z-score**: 데이터셋의 평균과 크게 다른 데이터 포인트를 식별하는 데 사용

## Z-score의 가정
- **Normal Distribution Assumption**: 데이터가 정상 분포(가우시안 분포)를 따른다고 가정
    - 데이터가 정상 분포에서 크게 벗어나면 부적절할 수 있음
- **Independence Assumption**: 데이터 포인트가 서로 독립적이라고 가정
    - 시계열 데이터나 공간 데이터는 이 가정을 위반할 수 있음
- **Outlier Proportion Assumption**: 데이터에서 이상치의 비율이 상대적으로 적다고 가정
    - 이상치가 많으면 성능이 저하될 수 있음
- **Equal Variance Assumption**: 데이터의 분산이 모든 수준에서 동일하다고 가정
    - 그룹이나 기간에 따라 분산이 크게 달라지면 부정확한 결과 초래
- **Symmetry Assumption**: 데이터 분포가 평균을 중심으로 대칭적이라고 가정
    - 데이터 분포가 비대칭적이거나 꼬리가 두꺼우면 부정확할 수 있음

## 장점
- **Simple Interpretation**: 이해와 구현이 쉬움
- **Effective for Normal Distributions**: 데이터가 정상 분포를 따를 때 효과적
- **Standardization**: 데이터를 표준화하여 다른 척도 및 단위 간 비교 가능
- **Quick Identification**: 평균에서 멀리 떨어진 데이터 포인트를 빠르게 식별

## 단점
- **Sensitive to Outliers**: 이상치에 민감하여 작은 데이터셋에서 허위 양성 가능성
- **Assumption of Normal Distribution**: 데이터가 정상 분포를 따르지 않으면 부정확
- **Unstable for Small Samples**: 작은 샘플에서는 신뢰할 수 없는 이상치 탐지
- **Global Distribution Impact**: 전체 데이터셋의 분포를 고려하여 지역 이상치 탐지에 부적절
- **Vulnerable to Skewness**: 데이터가 비대칭이면 진정한 이상치 식별이 어려움
- **Non-uniform Data**: 서로 다른 분포에서 온 데이터나 클러스터 밀도가 다른 경우 성능 저하

## 요약
- **Z-score**: 정상 분포 데이터를 위한 간단한 이상치 탐지 방법
- **주의사항**: 이상치에 민감하고 정상성 가정으로 인해 문제 발생 가능
- **데이터 특성 이해 필요**: Z-score를 사용하기 전에 데이터의 특성과 잠재적 단점을 고려해야 함


````python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Sample dataset
np.random.seed(42)  # 난수 생성을 위한 시드 값 설정 (재현 가능성 보장)
df = pd.DataFrame({  # 데이터프레임 생성
    'Feature1': np.random.normal(0, 1, 1000),  # 평균 0, 표준편차 1인 정규분포에서 1000개의 난수 생성
    'Feature2': np.random.normal(0, 1, 1000)   # 평균 0, 표준편차 1인 정규분포에서 1000개의 난수 생성
})

# Add some anomalies
df.loc[0] = [10, 10]  # 첫 번째 행에 이상치 [10, 10] 추가
df.loc[1] = [-5, -5]  # 두 번째 행에 이상치 [-5, -5] 추가 (쉼표 수정)

# Standardize the data
scaler = StandardScaler()  # StandardScaler 객체 생성
scaled_data = scaler.fit_transform(df)  # 데이터를 표준화 (평균 0, 표준편차 1로 변환)
scaled_data.shape  # (1000, 2) 출력

# Statistical Methods
# 1. Z-score
z_scores_anomalies = np.abs(scaled_data) > 3.5  # Z-score가 3.5를 초과하는 이상치 식별
print(z_scores_anomalies)

# 두 Feature 모두에서 이상치인 데이터 포인트를 식별
anomalies = np.any(z_scores_anomalies, axis=1)  # 각 행에서 하나라도 True인 경우 이상치로 간주
print(np.where(anomalies)[0])  # True인 인덱스 추출

# 데이터 시각화
plt.scatter(df['Feature1'], df['Feature2'], label='Normal Data')  # 정상 데이터 포인트 플로팅(이상치도 같이 플로팅하긴 함)
plt.scatter(df[anomalies]['Feature1'], df[anomalies]['Feature2'], color='red', label='Z-score Anomalies')  # 이상치 데이터 포인트 플로팅
plt.legend()  # 범례 추가
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('Anomaly Detection Results')  # 그래프 제목 추가
plt.show()  # 그래프 표시
````

# Modified Z-score anomalies (Statistical Method)

The modified Z-score is a variation of the Z-score method that uses the median and the median absolute deviation (MAD) to identify anomalies. It's a robust method that's less sensitive to outliers compared to the traditional Z-score.

## Modified Z-score Assumptions
- **No Assumption of Normal Distribution**: Modified Z-score는 데이터가 정상 분포를 따른다고 가정하지 않음. 따라서 정상성을 벗어나는 데이터에 적합함.
- **Independence Assumption**: 데이터 포인트가 서로 독립적이라고 가정. 이는 표준 Z-score와 동일.
- **Equal Variance Assumption**: 표준 Z-score만큼 분산 차이에 민감하지 않지만, 데이터셋 전반의 분산이 비교적 일관되면 더 좋음.
- **Outlier Proportion Assumption**: 이상치 비율이 상대적으로 큰 데이터셋에서도 유용. 극단값의 영향을 덜 받음.
- **Symmetry Assumption**: 비대칭 분포와 이상치에 더 강인하지만, 데이터의 분포와 대칭성을 평가하는 것이 여전히 중요함.

## 장점
1. **Robust to Outliers**: 표준 Z-score보다 이상치에 더 강함. 극단값의 영향을 덜 받는 중앙값과 MAD를 사용.
2. **No Assumption of Distribution**: 특정 데이터 분포를 가정하지 않아 다양한 데이터에 유연하게 적용 가능.
3. **Effective for Skewed Data**: 평균과 표준편차가 이상치에 의해 영향을 받을 수 있는 비대칭 데이터 분포에 유용.
4. **Simple Implementation**: 중앙값과 MAD 계산만으로 비교적 간단하게 구현 가능.

## 단점
1. **Threshold Selection**: 다른 임계값 기반 방법과 마찬가지로, 성능은 선택된 임계값에 크게 좌우됨. 적절한 임계값 선택을 위해 도메인 지식과 데이터 이해가 필요.
2. **Parameter Sensitivity**: 이상치에 덜 민감하지만, 임계값 선택 및 MAD 조정과 같은 매개변수 선택에 여전히 영향을 받음.
3. **Insensitive to Data Variability**: 중앙값과 MAD 외의 데이터 변동성을 고려하지 않음. 극단값이 아닌 이상치를 식별하는 데는 비효율적일 수 있음.
4. **Complex Anomalies**: 단순한 극단값이 아닌 독특한 패턴을 보이는 복잡한 이상치에 대해 성능이 떨어질 수 있음.
5. **Data Scaling**: 데이터에 따라, Modified Z-score를 적용하기 전에 데이터 스케일링이나 정규화가 필요할 수 있음.
6. **Contextual Information**: 다른 단변량 방법과 마찬가지로, 변수 간의 관계나 맥락 정보를 포착하지 못함.

Modified Z-score는 특히 비대칭 데이터 분포와 극단값이 포함된 데이터를 다룰 때 유용한 방법입니다. 그러나 임계값 선택 및 잠재적 매개변수 민감성과 같은 한계를 이해하고 데이터를 사용할 때의 맥락을 고려하는 것이 중요합니다.


````python
'''
Source: https://medium.com/analytics-vidhya/anomaly-detection-by-modified-z-score-f8ad6be62bac
Here we are going to make a slight modification and introduce a consistency correction k,
which allows us to use MAD as a consistent estimate for the standard deviation.
The value of k depends on the underlying distribution of the data. For simplicity, we will use k=1.4826.
'''

k = 1.4826  # MAD(Median Absolute Deviation)를 정규화하는 상수
threshold = 2.5  # 이상치를 판단하기 위한 Modified Z-score 임계값

# Modified Z-score 계산
modified_z_scores = (df - df.median()) / (k * (df - df.median()).abs().median())

# Modified Z-score가 임계값을 초과하는 이상치 식별
modified_z_scores_anomalies = np.where((modified_z_scores > threshold) | (modified_z_scores < -1 * threshold), True, False)

# 데이터 시각화
plt.scatter(df['Feature1'], df['Feature2'], label='Normal Data')  # 정상 데이터 포인트 플로팅
plt.scatter(df[modified_z_scores_anomalies]['Feature1'], df[modified_z_scores_anomalies]['Feature2'], color='red', label='Z-score Anomalies')  # 이상치 데이터 포인트 플로팅
plt.legend()  # 범례 추가
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('Anomaly Detection Results')  # 그래프 제목 추가
plt.show()  # 그래프 표시
````

# Percentile Anomalies - (Statistical Method)

Percentile 기반 이상치 탐지는 데이터 분포의 특정 **percentiles** 이하 또는 이상에 속하는 데이터 포인트를 식별하는 방법입니다. 간단한 방법이지만, 데이터를 적용하기 전에 다음과 같은 가정을 고려하는 것이 중요합니다:

## 가정
1. **Assumption of Data Distribution Shape**: Percentile 기반 이상치 탐지는 데이터 분포의 형태에 대한 가정을 하지 않습니다. 이는 비정상적이거나 비대칭 데이터를 처리할 때 유리할 수 있습니다.
2. **Independence Assumption**: 다른 많은 이상치 탐지 방법과 마찬가지로, Percentile 기반 탐지는 데이터 포인트가 서로 독립적이라고 가정합니다.
3. **Threshold Selection**: Percentile 기반 이상치 탐지의 효과는 선택된 **percentile thresholds**에 크게 좌우됩니다. 낮은 **percentile** (예: 1st 또는 5th **percentile**)은 비정상적으로 낮은 값을 식별하고, 높은 **percentile** (예: 95th 또는 99th **percentile**)은 비정상적으로 높은 값을 식별합니다. **percentiles** 선택은 도메인 지식과 데이터 특성에 의해 결정되어야 합니다.
4. **Outlier Proportion Assumption**: 이 방법은 데이터 내에 소수의 이상치가 존재하는 상황을 처리할 수 있습니다. 그러나 이상치가 데이터의 상당 부분을 차지하는 경우 Percentile 기반 탐지는 매우 효과적이지 않을 수 있습니다.
5. **Symmetry Assumption**: Percentile 기반 탐지는 대칭 및 비대칭 분포 모두에서 잘 작동할 수 있습니다. 이는 Z-score와 같은 방법에 비해 비대칭 데이터에 대해 더 강인합니다.
6. **Data Scaling**: 데이터와 사용 사례에 따라, Percentile 기반 이상치 탐지를 적용하기 전에 데이터를 스케일링하거나 정규화해야 할 수도 있습니다. 이는 **percentiles**이 의미 있게 계산되도록 합니다.

## 장점
1. **No Distribution Assumption**: Percentile 기반 탐지는 특정 데이터 분포를 가정하지 않습니다. 이는 다양한 분포 형태에 대해 강인하며, 비정상적이거나 비대칭 데이터를 처리하는 데 적합합니다.
2. **Simple and Intuitive**: 이 방법은 이해하기 쉽고 구현하기 간단합니다. **percentiles**을 기준으로 임계값을 설정하는 것은 쉽게 해석될 수 있습니다.
3. **Domain Knowledge**: Percentile 기반 탐지는 도메인 지식을 활용할 수 있습니다. 예를 들어, 특정 **percentile** 이하의 값이 이상치임을 알고 있는 경우, 그 정보를 효과적으로 사용할 수 있습니다.
4. **Effective for Skewed Data**: Percentile 기반 탐지는 Z-score와 같은 표준 편차 기반 방법이 어려움을 겪을 수 있는 비대칭 데이터 분포에 효과적입니다.

## 단점
1. **Threshold Selection**: Percentile 기반 탐지의 효과는 **threshold selection**에 크게 좌우됩니다. 적절한 **percentile**을 선택하려면 도메인 지식과 데이터 이해가 필요합니다.
2. **Outlier Proportion**: 데이터의 상당 부분이 이상치인 경우 이 방법의 성능이 저하될 수 있습니다. 이는 이상치 비율이 상대적으로 낮을 때 가장 적합합니다.
3. **Doesn't Capture Context**: Percentile 기반 탐지는 데이터의 지역적 맥락을 포착하지 않습니다. 데이터 포인트 간의 관계를 고려하지 않아 허위 양성이 발생할 수 있습니다.
4. **Insensitive to Location**: Percentile 기반 방법은 데이터의 위치나 추세를 고려하지 않습니다. 데이터의 대부분과 동일한 범위 내에 있는 이상치는 감지되지 않을 수 있습니다.
5. **Scaling Considerations**: 데이터 분포에 따라 **percentiles**을 정확하게 계산하기 위해 스케일링이나 정규화가 필요할 수 있습니다.
6. **Limited against Multivariate Data**: Percentile 기반 탐지는 다변량 데이터에 대해 덜 직관적입니다. 각 변수에 대한 **percentiles**을 정의해야 하며, 여러 변수가 상호작용할 때 이상치를 놓칠 수 있습니다.

![Percentiles](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/Standard_deviation_diagram.svg/325px-Standard_deviation_diagram.svg.png)


````python
# Set the percentiles for anomaly detection (e.g., 95th percentile)
percentile5 = 5  # 5th percentile 설정
percentile95 = 95  # 95th percentile 설정

# Calculate the threshold values based on percentiles for each feature
thresholds = {  # 각 Feature에 대한 percentile 기반 임계값 계산
    'Feature1_5': np.percentile(df['Feature1'], percentile5),  # Feature1의 5th percentile
    'Feature2_5': np.percentile(df['Feature2'], percentile5),  # Feature2의 5th percentile
    'Feature1_95': np.percentile(df['Feature1'], percentile95),  # Feature1의 95th percentile
    'Feature2_95': np.percentile(df['Feature2'], percentile95)  # Feature2의 95th percentile
}

# Detect anomalies based on the threshold values
anomaly_indices = (df['Feature1'] > thresholds['Feature1_95']) | (df['Feature2'] > thresholds['Feature2_95']) | (df['Feature1'] < thresholds['Feature1_5']) | (df['Feature2'] < thresholds['Feature2_5'])
# 각 Feature의 percentile 임계값을 기반으로 이상치 탐지
anomalies = df[anomaly_indices]  # 이상치 데이터 추출

# 데이터 시각화
plt.scatter(df['Feature1'], df['Feature2'], label='Normal Data')  # 정상 데이터 포인트 플로팅
plt.scatter(anomalies['Feature1'], anomalies['Feature2'], color='red', label='Percentile Anomalies')  # 이상치 데이터 포인트 플로팅
plt.legend()  # 범례 추가
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('Anomaly Detection Results')  # 그래프 제목 추가
plt.show()  # 그래프 표시
````

# Grubb's test

Grubbs' test는 단변량 데이터셋에서 이상치를 탐지하기 위한 통계적 방법입니다. 데이터 포인트의 값을 데이터셋의 평균과 표준편차와 비교하여 잠재적 이상치를 식별합니다. 이 테스트는 Grubbs' score 또는 G-score라는 통계량을 계산하여 데이터 포인트가 평균에서 표준편차 단위로 얼마나 떨어져 있는지를 정량화합니다.

Grubbs' score는 다음 공식으로 계산됩니다:

$G = \dfrac{|x - \bar{x}|}{s}$

여기서:
- **(G)**는 Grubbs' score입니다.
- **(x)**는 데이터 포인트의 값입니다.
- $\bar{x}$는 데이터셋의 평균입니다.
- **(s)**는 데이터셋의 표본 표준편차입니다.

Grubbs' score는 데이터 포인트가 평균에서 몇 표준편차 떨어져 있는지를 측정합니다. G-score가 클수록, 해당 데이터 포인트가 평균에서 멀리 떨어져 있을 가능성이 높아 이상치일 확률이 큽니다.

데이터 포인트가 이상치인지 여부를 결정하기 위해, G-score를 Grubbs' 분포 표의 임계값과 비교합니다. G-score가 임계값을 초과하면 해당 데이터 포인트는 이상치로 간주됩니다.

이 테스트는 반복적으로 수행될 수 있으며, 식별된 이상치를 데이터셋에서 제거하고 다시 테스트를 적용하여 더 이상 이상치가 감지되지 않을 때까지 반복할 수 있습니다.

Grubbs' test는 데이터가 정상 분포를 따른다고 가정하며, 데이터 포인트 수에 민감합니다. Grubbs' test는 단변량 데이터셋에서 개별 이상치를 식별하는 데 유용하지만, 다변량 또는 복잡한 데이터셋에는 다른 방법이 더 적합할 수 있습니다.

Grubbs' test는 단변량 데이터셋에서 이상치를 탐지하기 위해 널리 사용되는 통계적 방법입니다. 다른 기술과 마찬가지로 장단점이 있습니다. Grubbs' test의 장단점은 다음과 같습니다:

- Grubbs' test는 특히 데이터가 정상 분포를 따를 때 단변량 데이터셋에서 극단적인 이상치를 식별하는 데 유용하고 간단한 방법입니다.
- 그러나 정상성에서 벗어난 경우, 작은 샘플 크기, 그리고 덜 극단적이지만 여전히 중요한 이상치의 존재에 의해 효과가 제한될 수 있습니다. 데이터 분석 도구로서 유용하지만, 그 가정과 한계를 신중하게 고려해야 합니다.

**Pros:**

1. **Simple and Quick**: Grubbs' test는 이해하기 쉽고 구현하기 간단하여 이상치를 식별하는 빠른 초기 단계로 적합합니다.
2. **Univariate Focus**: 이 테스트는 단변량 데이터셋을 대상으로 하여 단일 변수의 분포를 분석할 때 적합합니다.
3. **Quantitative Measure**: Grubbs' test는 표준편차 단위로 평균에서의 편차 정도를 나타내는 정량적 점수(G-score)를 제공합니다.
4. **Identifies Extreme Values**: Grubbs' test는 평균에서 멀리 떨어진 개별 극단값을 식별하는 데 효과적입니다.
5. **Iterative Approach**: 이 테스트는 반복적으로 적용할 수 있으며, 이상치를 제거하고 재검사하여 이상치 탐지를 정밀하게 합니다.
6. **Assumes Normality**: Grubbs' test는 정상 분포 데이터를 대상으로 설계되었습니다. 데이터가 정상 분포를 따를 경우 매우 강력합니다.

**Cons:**

1. **Assumption of Normality**: Grubbs' test는 데이터가 정상 분포를 따른다고 가정합니다. 데이터 분포가 정상성을 크게 벗어나면 부정확한 결과를 초래할 수 있습니다.
2. **Univariate Limitation**: Grubbs' test는 단일 변수를 분석하는 데 제한적입니다. 여러 변수 간의 관계를 고려하지 않습니다.
3. **Sensitivity to Sample Size**: 테스트의 효과는 데이터 포인트 수에 의해 영향을 받을 수 있으며, 작은 샘플 크기에서는 잘 작동하지 않을 수 있습니다.
4. **Not Robust**: Grubbs' test는 극단값 자체에 민감합니다. 극단값이 평균과 표준편차에 영향을 미쳐 허위 양성을 초래할 수 있습니다.
5. **Limited to Extreme Outliers**: Grubbs' test는 개별 극단값을 탐지하도록 설계되었습니다. 극단적이지 않지만 여전히 중요한 이상치는 놓칠 수 있습니다.
6. **Outlier Definition Matters**: G-score의 임계값은 데이터 포인트 수와 선택한 유의 수준에 따라 달라집니다. "이상치"의 정의가 다르면 다른 결과를 초래할 수 있습니다.
7. **Multiple Testing Issues**: 큰 데이터셋이나 여러 변수를 대상으로 사용할 경우, Grubbs' test는 다중 검정으로 인한 허위 양성의 위험을 증가시킬 수 있습니다.
8. **Sensitive to Distribution Shape**: Grubbs' test는 분포의 형태에 민감할 수 있으며, 특히 꼬리가 두껍거나 정상성을 벗어난 경우 문제가 될 수 있습니다.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Grubbs' test를 사용한 이상치 탐지 함수 정의
def grubb_anomaly_detection(data, alpha=2.5):
    anomalies = []  # 이상치 인덱스를 저장할 리스트
    for col in data.columns:  # 각 컬럼에 대해 반복
        z_scores = np.abs(zscore(data[col]))  # 각 데이터 포인트의 z-score 계산
        mask = z_scores > alpha  # z-score가 alpha를 초과하는 데이터 포인트 찾기
        anomalies.extend(np.where(mask)[0])  # 이상치 인덱스를 anomalies 리스트에 추가
    return anomalies  # 이상치 인덱스 리스트 반환

# 테스트의 유의 수준(alpha) 설정
alpha = 2.5

# Grubbs' test를 사용한 이상치 탐지
anomalies = grubb_anomaly_detection(df, alpha=alpha)
print(anomalies)  # 이상치 인덱스 출력

print(df.iloc[anomalies])  # 이상치 데이터프레임 출력

# 이상치와 원래 데이터셋을 함께 플로팅
plt.scatter(df['Feature1'], df['Feature2'], label='Normal Data')  # 정상 데이터 포인트 플로팅
plt.scatter(df.iloc[anomalies]['Feature1'], df.iloc[anomalies]['Feature2'], color='red', label="Grubbs' test anomalies")  # 이상치 데이터 포인트 플로팅
plt.legend()  # 범례 추가
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('Anomaly Detection Results')  # 그래프 제목 추가
plt.show()  # 그래프 표시
````

# Local Outlier Factor (Density-Based Methods)

Local Outlier Factor (LOF)는 데이터 포인트의 이웃에 대한 지역 밀도 편차를 정량화하여 이상치를 탐지하는 알고리즘입니다. LOF는 데이터 포인트가 전체 데이터셋과 비교했을 때 지역 이웃 내에서 얼마나 고립되어 있는지를 측정하여 이상치를 식별합니다. LOF는 밀도가 불균일하거나 밀도가 다른 클러스터를 가진 데이터셋에서 특히 효과적입니다.

## Local Outlier Factor 알고리즘 작동 방식:

1. **Local Reachability Density (LRD) 계산:**
   - 각 데이터 포인트에 대해 LOF는 이웃에 대한 지역 도달 가능 밀도(LRD)를 계산합니다.
   - 이는 데이터 포인트에서 k-최근접 이웃까지의 평균 거리를 계산하여 이루어집니다 (k-distance).

2. **Local Outlier Factor (LOF) 계산:**
   - 그 후 LOF는 각 데이터 포인트에 대해 LOF를 계산하여 해당 포인트의 지역 이상치 정도를 정량화합니다.
   - 데이터 포인트의 LOF는 그 포인트의 LRD와 k-최근접 이웃의 LRD 비율의 평균입니다.
   - LOF 값이 높을수록 데이터 포인트가 이웃에 비해 저밀도 지역에 있음을 나타내며, 이는 이상치일 가능성을 시사합니다.

3. **이상치 식별:**
   - LOF 값을 기반으로 이상치를 식별합니다. LOF 값이 1보다 크게 나타나는 데이터 포인트는 이상치로 간주됩니다.
   - LOF 값이 1보다 작으면 데이터 포인트가 이웃에 비해 고밀도 지역에 있음을 나타냅니다.

4. **파라미터 튜닝:**
   - 고려할 최근접 이웃의 수를 나타내는 `k` 파라미터를 적절히 설정해야 합니다. 데이터셋의 특성에 따라 선택해야 합니다.

5. **확장성:**
   - LOF는 데이터 포인트 간 거리 계산이 포함되어 있어 큰 데이터셋에서는 계산 비용이 많이 들 수 있습니다.

LOF는 밀도와 형태가 다양한 클러스터를 가진 경우에 유리합니다. 또한, 노이즈 처리에 효과적이며 고차원 데이터에서도 작동할 수 있습니다. 그러나 파라미터 설정에 민감하며, 큰 데이터셋의 경우 계산 오버헤드가 있을 수 있습니다.

LOF를 사용할 때는 다양한 `k` 값을 실험하고 데이터 특성을 이해하는 것이 중요합니다. LOF는 전역적으로 분포된 이상치나 클러스터 밀도가 겹치는 경우에는 성능이 저하될 수 있습니다.

## LOF 알고리즘의 장단점:

**Pros:**

1. **Handles Varying Densities**: LOF는 밀도가 다른 클러스터나 불균일한 밀도를 가진 데이터셋에서 이상치를 효과적으로 식별합니다.
2. **Robust to Noise**: LOF는 노이즈에 상대적으로 강하며, 불규칙성이나 이상치가 있는 데이터를 잘 처리합니다.
3. **No Assumption of Data Distribution**: LOF는 특정 데이터 분포를 가정하지 않아 다양한 유형의 데이터에 적합합니다.
4. **Local Context**: LOF는 데이터 포인트의 지역 이웃에 대한 이상 상태를 평가하도록 설계되어, 지역 패턴과 클러스터에 민감합니다.
5. **Works in High Dimensions**: LOF는 고차원 데이터셋에서도 작동할 수 있어 복잡한 데이터 공간에서 이상치 탐지에 적합합니다.

**Cons:**

1. **Parameter Sensitivity**: LOF의 성능은 `k` 파라미터 (최근접 이웃 수)와 데이터 스케일링의 선택에 민감합니다. 적절한 `k` 값을 선택하는 것이 중요하며, 다른 `k` 값이 다른 결과를 초래할 수 있습니다.
2. **Computational Complexity**: LOF 계산은 데이터 포인트 간의 거리 계산을 포함하여 큰 데이터셋에서는 계산 비용이 많이 듭니다. 확장성을 위해 근사 기법이 필요할 수 있습니다.
3. **Overlapping Densities**: LOF는 클러스터 밀도가 겹치는 경우 이상치 탐지에 어려움을 겪을 수 있습니다.
4. **Global Outliers**: LOF는 주로 지역적 맥락에 중점을 두므로, 전역적으로 분포된 이상치 시나리오에서는 성능이 저하될 수 있습니다.
5. **Imbalanced Data**: LOF는 이상치가 정상 인스턴스에 비해 적다는 가정을 합니다. 데이터셋이 심하게 불균형한 경우 LOF의 성능이 영향을 받을 수 있습니다.
6. **Boundary Cases**: LOF는 클러스터 경계 근처의 이상치를 탐지하는 데 어려움을 겪을 수 있습니다.

요약하면, LOF는 밀도 클러스터와 지역 패턴이 있는 데이터셋에서 이상치를 탐지하는 데 강력한 알고리즘입니다. 그러나 파라미터 튜닝에 주의가 필요하며, 최적의 결과를 얻기 위해 도메인 지식과 탐색적 데이터 분석과 함께 사용해야 합니다.


````python
import pandas as pd
import numpy as np
from sklearn.neighbors import LocalOutlierFactor

# Convert the DataFrame to a NumPy array
X = df.values  # DataFrame을 NumPy 배열로 변환
# X shape : (1000, 2), numpy.ndarray

# Create LOF model
lof = LocalOutlierFactor(n_neighbors=5, contamination=0.01)  # LOF 모델 생성, 이웃 수를 5로 설정하고 오염 비율을 0.01로 설정

# Fit the model and predict anomalies
predictions = lof.fit_predict(X)  # 모델을 데이터에 적합시키고 이상치 예측
# preditions shape : (1000,), numpy.ndarray

# Get anomaly scores (negative LOF values)
anomaly_scores = -lof.negative_outlier_factor_  # 이상치 점수(음수 LOF 값)를 가져옴

# Identify anomalies
anomalies = df[predictions == -1]  # 예측 값이 -1인 데이터를 이상치로 식별

# 데이터 시각화
plt.scatter(df['Feature1'], df['Feature2'], color='blue', label='Normal')  # 정상 데이터 포인트 플로팅
plt.scatter(anomalies['Feature1'], anomalies['Feature2'], color='red', label='Anomalies')  # 이상치 데이터 포인트 플로팅

plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('LOF Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

# Isolation Forest - Machine Learning-Based Methods

**Isolation Forest** 알고리즘은 이상치 탐지를 위한 비지도 학습 기법입니다. 특히 고차원 데이터셋에서 이상치를 탐지하는 데 효율적입니다. 이 알고리즘은 이상치가 정상 데이터 포인트에 비해 분리하기 쉽다는 아이디어에 기반을 두고 있습니다. Isolation Forest는 데이터 공간을 더 작은 영역으로 분할하는 이진 트리의 앙상블을 구축합니다.

Isolation Forest는 데이터셋에서 이상치를 분리하는 효과적이고 효율적인 알고리즘입니다. 고차원 데이터셋에서 특히 유용하며, 전역 및 지역 이상치를 모두 처리할 수 있습니다. Isolation Forest의 기본 아이디어는 이상치(또는 아웃라이어)를 정상 데이터 포인트보다 더 빨리 분리하는 것입니다.

## Isolation Forest 알고리즘 작동 방식:

1. **Random Partitioning:**
   - 알고리즘은 임의로 특징을 선택한 후 해당 특징의 범위 내에서 분할 값을 무작위로 선택합니다.
   - 이 과정은 각 노드가 특징과 분할 값을 나타내는 이진 트리 구조를 생성합니다.

2. **Recursive Sub-sampling:**
   - 데이터 포인트는 재귀적으로 트리의 다른 가지로 분할됩니다. 적은 분할로 최종 노드에 도달하는 포인트는 이상치일 가능성이 큽니다.
   - 포인트를 분리하는 데 걸리는 분할 횟수는 "이상치 정도"의 척도로 간주될 수 있습니다.

3. **Scoring Anomalies:**
   - 데이터 포인트가 트리를 통해 이동하는 동안 평균 경로 길이를 계산합니다.
   - 이상치는 일반적으로 더 짧은 평균 경로 길이를 가지며, 이는 더 빨리 분리되기 때문입니다. 정상 데이터 포인트는 분리하는 데 더 많은 분할이 필요합니다.

4. **Anomaly Detection:**
   - 알고리즘은 각 데이터 포인트에 평균 경로 길이를 기반으로 이상치 점수를 할당합니다.
   - 짧은 평균 경로 길이(낮은 값)를 가진 데이터 포인트는 이상치로 간주됩니다.

## Isolation Forest의 장점:

1. **Effective for High Dimensions:** Isolation Forest는 고차원 데이터셋에서 효과적이며, 전통적인 거리 기반 방법이 어려움을 겪는 경우에도 잘 작동합니다.
2. **Efficiency:** 알고리즘의 효율성 덕분에 대규모 데이터셋에 적합합니다. 많은 메모리를 필요로 하지 않으며 병렬화가 가능합니다.
3. **Handles Clusters:** Isolation Forest는 클러스터 내의 이상치를 식별할 수 있어, 밀도가 다른 영역을 가진 데이터셋에 적합합니다.
4. **Scalability:** Isolation Forest는 성능 저하 없이 대규모 데이터셋으로 확장할 수 있습니다.

## Isolation Forest의 단점:

1. **Parameter Sensitivity:** 알고리즘의 성능은 숲의 나무 수와 기타 하이퍼파라미터에 따라 달라질 수 있습니다.
2. **Randomness:** 특징과 분할 값을 선택할 때의 무작위성이 결과의 변동성을 초래할 수 있습니다.
3. **Global Outliers:** Isolation Forest는 빠르게 분리되지 않는 전역 이상치를 탐지하는 데 최선의 선택이 아닐 수 있습니다.
4. **Data Scaling:** Isolation Forest는 데이터 스케일링을 본질적으로 처리하지 않습니다. 스케일링은 특히 고차원 공간에서 결과에 영향을 미칠 수 있습니다.

요약하면, Isolation Forest 알고리즘은 특히 고차원 데이터셋에서 이상치를 탐지하는 강력하고 효율적인 방법입니다. 클러스터와 지역 이상치를 처리하는 능력 덕분에 다용도로 사용할 수 있습니다. 그러나 어떤 알고리즘과 마찬가지로 파라미터 튜닝과 데이터 및 사용 사례에 따른 강점과 한계를 고려해야 합니다.


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Convert the DataFrame to a NumPy array
X = df.values  # DataFrame을 NumPy 배열로 변환

# Create Isolation Forest model
clf = IsolationForest(contamination=0.1, random_state=42)  # Isolation Forest 모델 생성, 오염 비율을 0.1로 설정하고 랜덤 시드 값을 42로 설정

# Fit the model and predict anomalies
predictions = clf.fit_predict(X)  # 모델을 데이터에 적합시키고 이상치 예측

# Identify anomalies
anomalies = X[predictions == -1]  # 예측 값이 -1인 데이터를 이상치로 식별
normal_points = X[predictions == 1]  # 예측 값이 1인 데이터를 정상 데이터로 식별

# anomalies와 normal_points의 자료형과 shape 출력
print(f"anomalies 자료형: {type(anomalies)}, shape: {anomalies.shape}")
print(f"normal_points 자료형: {type(normal_points)}, shape: {normal_points.shape}")

# 데이터와 이상치를 플로팅
plt.figure(figsize=(8, 6))  # 그림 크기 설정
plt.scatter(normal_points[:, 0], normal_points[:, 1], color='blue', label='Normal')  # 정상 데이터 포인트 플로팅
plt.scatter(anomalies[:, 0], anomalies[:, 1], color='red', label='Anomaly')  # 이상치 데이터 포인트 플로팅
plt.xlabel('Feature 1')  # x축 라벨 추가
plt.ylabel('Feature 2')  # y축 라벨 추가
plt.title('Isolation Forest Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

# DBSCAN - Density-Based Method

DBSCAN (Density-Based Spatial Clustering of Applications with Noise)는 주로 클러스터링 알고리즘이지만, 클러스터에 할당되지 않은 점을 이상치로 간주함으로써 이상치 탐지에도 사용할 수 있습니다. DBSCAN은 데이터의 밀집 영역을 식별하고, 희박한 영역의 점을 노이즈 또는 이상치로 간주합니다.

요약하면, DBSCAN은 주로 클러스터링을 위해 설계되었지만, 노이즈 점을 이상치로 간주하여 이상치 탐지에도 사용할 수 있습니다. 이는 밀집 클러스터 내의 지역 이상치 탐지에 가장 효과적이지만, 의미 있는 결과를 얻기 위해서는 파라미터 튜닝과 데이터 특성에 대한 고려가 중요합니다.

## DBSCAN 알고리즘 작동 방식:

1. **Density-Based Clustering:**
   - DBSCAN은 데이터셋에서 무작위로 한 점을 선택하는 것으로 시작합니다.
   - 그런 다음 시작점에서 지정된 거리(엡실론) 내의 모든 점을 식별합니다. 이 점들은 이웃으로 간주됩니다.

2. **Core Points and Density Reachability:**
   - 점 주변의 이웃 수가 지정된 최소 점 수(min_samples)보다 많으면, 그 점은 "코어 포인트"로 간주됩니다.
   - 각 코어 포인트에 대해 DBSCAN은 엡실론 거리 내에서 직접 또는 간접적으로 도달 가능한 모든 점을 식별합니다. 이들은 클러스터를 형성합니다.

3. **Noise Points:**
   - 코어 포인트가 아니거나 어떤 코어 포인트에서도 도달할 수 없는 점들은 노이즈 포인트 또는 이상치로 간주됩니다.

4. **Anomaly Detection using Noise:**
   - 이상치 탐지의 맥락에서 DBSCAN이 식별한 노이즈 포인트를 이상치로 간주할 수 있습니다.
   - 이는 어떤 밀집 클러스터에도 잘 맞지 않는 데이터 포인트들입니다.

## DBSCAN을 이용한 이상치 탐지의 장점:

1. **Cluster-Based Approach:** DBSCAN은 밀집 클러스터 외부에 있는 이상치를 식별하는 데 효과적이며, 지역 이상치 탐지에 유용합니다.
2. **Adaptive to Density:** DBSCAN은 다양한 데이터 밀도에 잘 적응하여, 비균일 밀도 영역을 가진 데이터셋에 적합합니다.
3. **No Assumption of Data Distribution:** DBSCAN은 특정 데이터 분포를 가정하지 않습니다.

## DBSCAN을 이용한 이상치 탐지의 단점:

1. **Threshold Selection:** 엡실론과 min_samples의 적절한 값을 설정하는 것이 효과적인 이상치 탐지에 필수적입니다. 잘못된 선택은 이상치를 놓치거나 정상 포인트를 이상치로 분류하는 결과를 초래할 수 있습니다.

3. **Sensitive to Parameters:** 성능은 파라미터 선택에 민감하며, 최적의 값을 찾는 것이 어려울 수 있습니다.

# kNN - Distance-Based Methods

k-Nearest Neighbors (kNN)는 데이터 포인트가 이웃과 얼마나 유사한지를 고려하여 이상치를 탐지할 수 있는 다용도 알고리즘입니다. 이상치는 정상 데이터 포인트에 비해 유사한 이웃의 수가 적을 것입니다. kNN을 이용한 이상치 탐지 방법은 다음과 같습니다:

## kNN을 이용한 이상치 탐지 방법:

1. **Select k:**
   - 고려할 최근접 이웃의 수인 k 값을 선택합니다. 적절한 k 값은 데이터에 따라 달라지며 신중하게 선택해야 합니다. k 값이 작으면 노이즈에 민감하고, k 값이 크면 지역적 이상치를 놓칠 수 있습니다.

2. **Calculate Distances:**
   - 관심 있는 데이터 포인트와 데이터셋의 모든 다른 데이터 포인트 간의 거리를 계산합니다.

3. **Find k Nearest Neighbors:**
   - 대상 데이터 포인트와 거리가 가장 짧은 k개의 데이터 포인트를 식별합니다.

4. **Calculate Average Distance:**
   - k개의 최근접 이웃으로부터 대상 데이터 포인트까지의 평균 거리를 계산합니다.

5. **Anomaly Detection:**
   - k개의 최근접 이웃의 평균 거리가 다른 대부분의 데이터 포인트의 평균 거리보다 크게 나타나면, 해당 데이터 포인트는 이상치로 간주됩니다.

## kNN을 이용한 이상치 탐지의 장점:

1. **Adaptive to Data:** kNN은 다양한 유형의 데이터에 적응하며 특정 분포를 가정하지 않습니다.
2. **Local Anomaly Detection:** kNN은 밀집 영역 내의 지역적 이상치를 탐지하는 데 특히 효과적입니다.
3. **Flexibility:** kNN은 다양한 형태, 밀도 및 차원을 가진 데이터를 처리할 수 있습니다.
4. **No Assumption of Data Distribution:** kNN은 특정 데이터 분포를 가정하지 않기 때문에 다양한 데이터 유형에 적용할 수 있습니다.

## kNN을 이용한 이상치 탐지의 단점:

1. **Parameter Sensitivity:** k 값의 선택이 결과에 크게 영향을 미칩니다. 부적절한 k 값은 허위 양성 또는 음성을 초래할 수 있습니다.
2. **Distance Metric:** 거리 측정 기준의 선택이 중요합니다. 데이터에 따라 유클리드 거리가 최적이 아닐 수 있으며, 스케일링과 정규화가 중요합니다.
3. **Curse of Dimensionality:** kNN의 성능은 고차원 공간에서 저하될 수 있습니다. 차원이 클수록 거리 기반 방법은 어려움을 겪습니다.
4. **Global Anomalies:** kNN은 모든 데이터 포인트와 크게 다른 전역 이상치를 탐지하는 데 효과적이지 않을 수 있습니다.
5. **Computational Complexity:** kNN은 모든 데이터 포인트에 대한 거리를 계산해야 하므로, 큰 데이터셋에서는 계산 비용이 많이 들 수 있습니다.

요약하면, kNN은 밀집 영역 내에서 지역적 이상치를 탐지하는 데 특히 적합한 유연하고 적응력 있는 방법입니다. 그러나 의미 있는 결과를 얻기 위해서는 신중한 파라미터 튜닝과 거리 측정 기준에 대한 주의가 필요합니다.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

# Create kNN model
n_neighbors = 4  # 고려할 최근접 이웃의 수 설정
knn = NearestNeighbors(n_neighbors=n_neighbors)  # kNN 모델 생성
knn.fit(df)  # 모델을 데이터에 적합시킴

# Calculate distances and indices of k-nearest neighbors
distances, indices = knn.kneighbors()  # k-최근접 이웃의 거리와 인덱스 계산

# Calculate the average distance to k-nearest neighbors for each point
avg_distances = np.mean(distances, axis=1)  # 각 포인트에 대한 k-최근접 이웃까지의 평균 거리 계산
avg_distances.shape  # (1000,)
print(type(avg_distances))


# Set a threshold for anomaly detection (you can adjust this as needed)
distance_threshold = np.percentile(avg_distances, 95)  # 이상치 탐지를 위한 임계값 설정 (95번째 백분위수)
# 상위 5%의 평균 거리보다 큰 값들은 이상치로 간주
print(type(distance_threshold))

# Identify anomalies
anomalies = df[avg_distances > distance_threshold]  # 평균 거리가 임계값을 초과하는 포인트를 이상치로 식별

# Plot the dataset and anomalies
plt.scatter(df['Feature1'], df['Feature2'], label='Normal')  # 정상 데이터 포인트 플로팅
plt.scatter(anomalies['Feature1'], anomalies['Feature2'], color='red', label='Anomalies')  # 이상치 데이터 포인트 플로팅
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('kNN Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

# Mahalanobis Distance - Distance Based Method

Mahalanobis Distance는 변수들 간의 상관관계를 고려하여 다변량 공간에서 데이터 포인트와 분포 간의 거리를 정량화하는 측정 방법입니다. 이는 특히 상관된 특징을 가진 데이터를 다룰 때 이상치 탐지에 자주 사용됩니다. Mahalanobis Distance를 이용한 이상치 탐지 방법은 다음과 같습니다:

## Mahalanobis Distance를 이용한 이상치 탐지 방법:

1. **Calculate Covariance Matrix:**
   - 데이터셋의 공분산 행렬을 계산합니다. 이 행렬은 변수들 간의 관계를 나타내며, 변수들이 함께 얼마나 변하는지를 나타냅니다.

2. **Calculate Mean Vector:**
   - 데이터셋의 평균 벡터를 계산합니다. 이 벡터는 각 변수의 평균 값을 나타냅니다.

3. **Calculate Inverse Covariance Matrix:**
   - 공분산 행렬의 역행렬을 계산합니다. 이는 Mahalanobis Distance를 계산하는 데 필요합니다.


4. **Calculate Mahalanobis Distance:**
   - For each data point, calculate the Mahalanobis Distance using the formula:
     ```
     MD = sqrt((x - μ)^T * Σ^-1 * (x - μ))
     ```
     where:
     - `x` is the data point.
     - `μ` is the mean vector.
     - `Σ^-1` is the inverse covariance matrix.

5. **Threshold and Anomaly Detection:**
   - Mahalanobis Distance에 대한 임계값을 설정합니다. 이 거리 임계값을 초과하는 데이터 포인트는 이상치로 간주됩니다.

## Mahalanobis Distance를 이용한 이상치 탐지의 장점:

1. **Accounts for Correlation:** Mahalanobis Distance는 변수들 간의 상관관계를 고려하므로, 상관된 특징을 가진 데이터셋에서 이상치를 탐지하는 데 특히 유용합니다.
2. **Multivariate Analysis:** Mahalanobis Distance는 다변량 데이터에서 이상치를 탐지하는 데 적합합니다.
3. **No Assumption of Normal Distribution:** Mahalanobis Distance는 특정 데이터 분포를 가정하지 않습니다.
4. **Effective for Elliptical Clusters:** Mahalanobis Distance는 타원형 클러스터에서 이상치를 탐지하는 데 효과적입니다.

## Mahalanobis Distance를 이용한 이상치 탐지의 단점:

1. **Sensitivity to Multivariate Outliers:** Mahalanobis Distance는 다변량 이상치에 민감할 수 있으며, 이는 공분산 행렬에 강하게 영향을 미칠 수 있습니다.
2. **Parameter Selection:** Mahalanobis Distance에 대한 임계값을 신중하게 선택해야 하며, 이는 어려울 수 있습니다.
3. **Computationally Intensive:** 공분산 행렬의 역행렬을 계산하고 Mahalanobis Distance를 계산하는 것은 큰 데이터셋에 대해 계산 비용이 많이 들 수 있습니다.
4. **Scaling Consideration:** Mahalanobis Distance는 데이터의 스케일에 민감하므로 변수에 대한 스케일링과 정규화가 필요할 수 있습니다.

요약하면, Mahalanobis Distance는 상관된 특징을 가진 다변량 데이터에서 이상치를 탐지하는 데 유용한 도구입니다. 그러나 의미 있는 결과를 얻기 위해서는 파라미터 선택과 데이터 특성에 대한 신중한 고려가 필요합니다.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import mahalanobis

# 데이터의 평균 벡터와 공분산 행렬 계산
mean_vector = df.mean()  # 평균 벡터 계산
mean_vector.shape  # (2,)
print(mean_vector)
cov_matrix = df.cov()  # 공분산 행렬 계산
cov_matrix.shape  # (2, 2)
print(cov_matrix)

# 공분산 행렬의 역행렬 계산
inv_cov_matrix = np.linalg.inv(cov_matrix)  # 공분산 행렬의 역행렬 계산
print(inv_cov_matrix)

# 각 데이터 포인트에 대한 Mahalanobis Distance 계산
mahalanobis_distances = []  # Mahalanobis Distance를 저장할 리스트 초기화
for _, row in df.iterrows():  # 데이터 프레임의 각 행에 대해 반복
    mahalanobis_distances.append(mahalanobis(row, mean_vector, inv_cov_matrix))  # Mahalanobis Distance 계산 후 리스트에 추가

print(len(mahalanobis_distances)) # 1000
mahalanobis_distances = np.array(mahalanobis_distances)  # 리스트를 numpy 배열로 변환

# 이상치 탐지를 위한 Mahalanobis Distance 임계값 설정
mahalanobis_distance_threshold = 2.5  # 임계값 설정

# 이상치의 인덱스 찾기
anomaly_indices = np.where(mahalanobis_distances > mahalanobis_distance_threshold)[0]  # Mahalanobis Distance가 임계값을 초과하는 인덱스 찾기
type(anomaly_indices)  # <class 'numpy.ndarray'>
print(anomaly_indices)

# 데이터셋을 플로팅하고 이상치 강조 표시
plt.scatter(df['Feature1'], df['Feature2'], label='Normal')  # 정상 데이터 포인트 플로팅
plt.scatter(df.iloc[anomaly_indices]['Feature1'], df.iloc[anomaly_indices]['Feature2'], color='red', label='Anomaly')  # 이상치 데이터 포인트 플로팅
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('Anomaly Detection using Mahalanobis Distance')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

# One-Class SVM - Machine Learning-Based Methods

A One-Class SVM (Support Vector Machine)는 이상치 탐지에 사용되는 기계 학습 알고리즘입니다. 전통적인 SVM은 분류 작업에 사용되지만, One-Class SVM은 라벨이 없는 데이터에서 이상치 또는 이상점을 학습하고 식별하도록 설계되었습니다. 이 알고리즘은 정상 데이터의 분포를 학습하고, 이 분포에서 크게 벗어나는 인스턴스를 이상치로 식별합니다.

One-Class SVM은 이진 분류에서 사용하는 초평면(hyperplane)을 사용하는 대신, 모든 인스턴스를 포함하는 초구(hypersphere)를 사용합니다. 여기서 "가장 큰 가능성 있는 마진"을 "가장 작은 가능성 있는 초구"로 생각할 수 있습니다.

요약하면, One-Class SVM은 불균형 데이터와 고차원 공간을 처리할 때 특히 강력한 이상치 탐지 기법입니다. 하지만 정확한 결과를 얻기 위해서는 하이퍼파라미터 튜닝과 데이터 특성에 대한 신중한 고려가 필요합니다.

## One-Class SVM을 이용한 이상치 탐지 방법:

1. **Training the Model:**
   - One-Class SVM은 정상 데이터만으로 학습되며, 데이터셋에서 이상치가 소수임을 가정합니다.
   - 정상 데이터를 알고리즘에 제공하고 커널(예: Gaussian 또는 Radial Basis Function)을 지정합니다.

2. **Finding the Separating Hyperplane:**
   - One-Class SVM의 목표는 정상 데이터 포인트를 포함하고 이상치를 배제하는 초평면을 찾는 것입니다.

3. **Anomaly Detection:**
   - 테스트 중에는 초평면의 정상 데이터 포인트가 적은 쪽에 속하는 데이터 포인트가 이상치로 분류됩니다.

## One-Class SVM을 이용한 이상치 탐지의 장점:

1. **Flexibility:** One-Class SVM은 고차원 공간의 데이터를 처리할 수 있으며, 다양한 유형의 데이터 분포에 적합합니다.
2. **Handles Unbalanced Data:** One-Class SVM은 데이터셋에서 이상치가 소수일 때 효과적입니다.
3. **Non-linear Separation:** One-Class SVM은 정상 데이터와 이상치 간의 비선형 경계를 포착할 수 있습니다.
4. **Robust to Noise:** One-Class SVM은 훈련 데이터의 노이즈에 강하며, 정상 데이터의 전반적인 패턴을 찾는 데 중점을 둡니다.

## One-Class SVM을 이용한 이상치 탐지의 단점:

1. **Hyperparameter Tuning:** 적절한 커널을 선택하고 하이퍼파라미터를 튜닝하는 것이 어렵고 교차 검증이 필요합니다.
2. **Scalability:** One-Class SVM의 성능은 계산 복잡성 때문에 큰 데이터셋에서 저하될 수 있습니다.
3. **Global Anomalies:** One-Class SVM은 정상 데이터와 명확한 경계를 가지지 않는 전역 이상치를 탐지하는 데 효과적이지 않을 수 있습니다.
4. **Sensitive to Outliers:** One-Class SVM은 훈련 데이터에 이상치가 있을 경우 민감하게 반응할 수 있습니다.
5. **Data Scaling:** 비선형 커널을 사용할 때는 스케일링과 정규화가 필요할 수 있습니다.
6. **Interpretability:** 초평면의 해석이 어렵기 때문에 이상치의 특성을 이해하기 어려울 수 있습니다.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import OneClassSVM

# 데이터를 NumPy 배열로 변환
data = df.values  # DataFrame을 NumPy 배열로 변환

# One-Class SVM 모델 생성
nu_value = 0.05  # 이상치 비율 설정
ocsvm = OneClassSVM(kernel='rbf', nu=nu_value)  # RBF 커널을 사용하는 One-Class SVM 모델 생성
# RBF 커널은 고차원 공간으로 매핑하여 비선형 데이터를 선형적으로 분리 가능

# 모델 학습
ocsvm.fit(data)  # 데이터를 사용하여 모델 학습

# 정상 데이터와 이상치 예측
predictions = ocsvm.predict(data)  # 데이터를 사용하여 예측
inliers = data[predictions == 1]  # 정상 데이터 (1로 예측된 데이터)
outliers = data[predictions == -1]  # 이상치 (-1로 예측된 데이터)

# 데이터와 이상치 플로팅
plt.scatter(data[:, 0], data[:, 1], color='blue', label='Data')  # 모든 데이터 포인트 플로팅
plt.scatter(outliers[:, 0], outliers[:, 1], color='red', label='Anomalies')  # 이상치 데이터 포인트 플로팅
plt.xlabel('Feature 1')  # x축 라벨 추가
plt.ylabel('Feature 2')  # y축 라벨 추가
plt.title('One-Class SVM Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

# Autoencoder - Machine Learning technique

Autoencoder를 사용한 이상치 탐지는 Autoencoder 신경망을 학습시켜 정상 데이터에 존재하는 패턴을 학습하는 것을 포함합니다. 그런 다음 Autoencoder가 새로운 데이터 샘플을 얼마나 잘 재구성하는지를 관찰하여 이상치를 탐지합니다. Autoencoder가 샘플을 정확하게 재구성하는 데 어려움을 겪을 때 이상치가 존재할 가능성이 있습니다.

요약하면, Autoencoder는 데이터가 고차원적이고 복잡한 패턴을 가진 경우 이상치 탐지에 효과적일 수 있습니다. 그러나 적절한 튜닝과 평가가 필요하며, 아키텍처의 성능은 특정 데이터셋과 이상치 특성에 따라 평가되어야 합니다.

Autoencoder는 차원 축소 및 특징 학습에 일반적으로 사용되는 신경망 아키텍처입니다. 고차원 데이터를 처리할 때 특히 유용합니다. Autoencoder를 사용한 이상치 탐지 방법은 다음과 같습니다:

## Autoencoder를 사용한 이상치 탐지 방법:

1. **Architecture Setup:**
   - Autoencoder는 인코더와 디코더로 구성됩니다. 인코더는 입력 데이터를 낮은 차원의 잠재 공간 표현으로 압축하고, 디코더는 이 표현으로부터 원래 입력을 재구성합니다.

2. **Training on Normal Data:**
   - Autoencoder를 정상 데이터만을 사용하여 학습시킵니다. 네트워크는 정상 데이터 패턴을 재구성하는 방법을 학습합니다.

3. **Reconstruction Loss:**
   - 학습 중에는 Autoencoder의 입력과 출력 간의 차이인 재구성 손실이 계산됩니다. 이 손실은 원래 데이터와 재구성된 데이터 간의 불일치를 캡처합니다.

4. **Anomaly Detection:**
   - 학습 후, 정상 데이터와 이상치를 모두 Autoencoder에 입력합니다.
   - 각 데이터 포인트에 대해 재구성 손실을 계산합니다.
   - 재구성 손실이 높은 데이터 포인트는 학습 중에 학습된 패턴에서 크게 벗어나기 때문에 이상치일 가능성이 큽니다.

## Autoencoder를 사용한 이상치 탐지의 장점:

1. **Unsupervised Learning:** Autoencoder는 비지도 학습 방식으로 학습될 수 있으며, 학습을 위해 정상 데이터만 필요합니다.
2. **High-Dimensional Data:** Autoencoder는 고차원 데이터를 처리할 수 있으며, 데이터 내의 복잡한 관계를 캡처할 수 있습니다.
3. **Non-linear Patterns:** Autoencoder는 비선형 패턴을 캡처할 수 있어 다양한 유형의 데이터에서 이상치 탐지에 적합합니다.
4. **Feature Extraction:** Autoencoder는 데이터로부터 의미 있는 특징을 학습하여 이상치 식별에 도움을 줍니다.

## Autoencoder를 사용한 이상치 탐지의 단점:

1. **False Positives:** 이상치가 특정 정상 데이터 패턴과 유사한 경우 Autoencoder는 거짓 양성을 생성할 수 있습니다.
2. **Hyperparameter Tuning:** 아키텍처 크기, 학습률, 학습 에포크 수 등 최적의 성능을 위해 하이퍼파라미터 튜닝이 필요합니다.
3. **Limited Interpretability:** 이상치로 표시된 인스턴스가 왜 그런지 이해하기 어려울 수 있어 해석 가능성이 제한될 수 있습니다.
4. **Training Data Quality:** Autoencoder의 성능은 학습 데이터의 품질에 크게 의존합니다. 노이즈가 많거나 편향된 데이터는 결과에 영향을 미칠 수 있습니다.
5. **Class Imbalance:** Autoencoder는 주로 정상 데이터에서 학습하기 때문에 클래스 불균형 문제에 어려움을 겪을 수 있습니다.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.layers import Input, Dense
from keras.models import Model
from sklearn.preprocessing import StandardScaler

# 데이터 표준화
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)  # 데이터를 표준화

# 재현성을 위한 랜덤 시드 설정
np.random.seed(42)

# Autoencoder 모델 구성
input_layer = Input(shape=(2,))  # 입력 레이어 (특징의 수는 2)
encoded = Dense(1, activation='relu')(input_layer)  # 인코딩 레이어 (잠재 공간 크기는 1)
decoded = Dense(2)(encoded)  # 디코딩 레이어 (출력 크기는 2)

autoencoder = Model(inputs=input_layer, outputs=decoded)  # Autoencoder 모델 정의
autoencoder.compile(optimizer='adam', loss='mean_squared_error')  # 모델 컴파일 (최적화기: Adam, 손실 함수: MSE)

# Autoencoder 학습
autoencoder.fit(scaled_data, scaled_data, epochs=20, batch_size=100, verbose=0)  # 모델 학습

# 재구성 오차 계산
reconstructed_data = autoencoder.predict(scaled_data)  # 재구성된 데이터 예측
print(type(reconstructed_data))  # <class 'numpy.ndarray'>

# 재구성된 데이터의 shape 확인
print("Shape of Reconstructed Data:", reconstructed_data.shape)  # (1000, 2)

# 재구성된 데이터의 일부 샘플 출력
print("Reconstructed Data Samples:")
print(reconstructed_data[:3])  # 재구성된 데이터의 첫 3개 샘플 출력

reconstruction_errors = np.mean(np.square(scaled_data - reconstructed_data), axis=1)  # 재구성 오차 계산 (MSE)
# reconstruction_errors.shape()  # (1000,)
print(type(reconstruction_errors))  # <class 'numpy.ndarray'>
# 재구성 오차의 첫 3개 값 출력
print("Reconstruction Errors Samples:")
print(reconstruction_errors[:3])

# 이상치 탐지를 위한 재구성 오차 임계값 설정
reconstruction_error_threshold = np.percentile(reconstruction_errors, 95)  # 재구성 오차의 95번째 백분위수

# 이상치 식별
anomalies = df[reconstruction_errors > reconstruction_error_threshold]  # 재구성 오차가 임계값을 초과하는 데이터 식별

# 데이터셋과 이상치 플로팅
plt.scatter(df['Feature1'], df['Feature2'], label='Normal')  # 정상 데이터 포인트 플로팅
plt.scatter(anomalies['Feature1'], anomalies['Feature2'], color='red', label='Anomalies')  # 이상치 데이터 포인트 플로팅
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('Autoencoder Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

## PyTorch Version

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler

# 데이터 표준화
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)  # 데이터를 표준화

# 재현성을 위한 랜덤 시드 설정
np.random.seed(42)
torch.manual_seed(42)

# Autoencoder 모델 구성
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(2, 1),  # 입력 크기: 2, 출력 크기: 1
            nn.ReLU(True)
        )
        self.decoder = nn.Sequential(
            nn.Linear(1, 2)  # 입력 크기: 1, 출력 크기: 2
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# 모델 인스턴스 생성
autoencoder = Autoencoder()

# 손실 함수와 최적화기 정의
criterion = nn.MSELoss()
optimizer = optim.Adam(autoencoder.parameters(), lr=0.001)

# 데이터를 Tensor로 변환
scaled_data_tensor = torch.tensor(scaled_data, dtype=torch.float32)

# Autoencoder 학습
num_epochs = 20
batch_size = 100

for epoch in range(num_epochs):
    # 매 epoch마다 데이터셋을 무작위로 섞기 위해 인덱스를 랜덤하게 생성
    permutation = torch.randperm(scaled_data_tensor.size()[0])
    for i in range(0, scaled_data_tensor.size()[0], batch_size):
        # 배치 단위로 데이터 인덱스 선택
        indices = permutation[i:i + batch_size]
        batch_x = scaled_data_tensor[indices]

        # Forward pass: Autoencoder를 통해 배치 데이터를 입력하고 출력값 계산
        outputs = autoencoder(batch_x)
        # 손실 함수(MSE)를 사용하여 원본 데이터(batch_x)와 출력값(outputs) 간의 손실 계산
        loss = criterion(outputs, batch_x)

        # Backward and optimize
        optimizer.zero_grad()  # 옵티마이저의 기존 그래디언트 값을 초기화
        loss.backward()  # 역전파를 통해 그래디언트 계산
        optimizer.step()  # 옵티마이저를 사용하여 파라미터 업데이트

# 재구성 오차 계산
reconstructed_data = autoencoder(scaled_data_tensor).detach().numpy()  # 재구성된 데이터 예측
print(type(reconstructed_data))  # <class 'numpy.ndarray'>

# 재구성된 데이터의 shape 확인
print("Shape of Reconstructed Data:", reconstructed_data.shape)  # (1000, 2)

# 재구성된 데이터의 일부 샘플 출력
print("Reconstructed Data Samples:")
print(reconstructed_data[:3])  # 재구성된 데이터의 첫 3개 샘플 출력

reconstruction_errors = np.mean(np.square(scaled_data - reconstructed_data), axis=1)  # 재구성 오차 계산 (MSE)
print(type(reconstruction_errors))  # <class 'numpy.ndarray'>

# 재구성 오차의 첫 3개 값 출력
print("Reconstruction Errors Samples:")
print(reconstruction_errors[:3])

# 이상치 탐지를 위한 재구성 오차 임계값 설정
reconstruction_error_threshold = np.percentile(reconstruction_errors, 95)  # 재구성 오차의 95번째 백분위수

# 이상치 식별
anomalies = df[reconstruction_errors > reconstruction_error_threshold]  # 재구성 오차가 임계값을 초과하는 데이터 식별

# 데이터셋과 이상치 플로팅
plt.scatter(df['Feature1'], df['Feature2'], label='Normal')  # 정상 데이터 포인트 플로팅
plt.scatter(anomalies['Feature1'], anomalies['Feature2'], color='red', label='Anomalies')  # 이상치 데이터 포인트 플로팅
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('Autoencoder Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

# Moving Averages (time series approach)

Moving Averages(이동 평균)는 시계열 데이터에서 이상치를 탐지하기 위한 간단하면서도 효과적인 기법입니다.

요약하면, 이동 평균은 시계열 데이터에서 이상치를 탐지하기 위한 간단하고 빠른 방법입니다. 장기적인 경향과 패턴을 식별하는 데 특히 효과적이지만, 최근 이상치를 탐지하는 데는 지연이 발생할 수 있으며, 급격한 변화나 복잡한 패턴이 있는 데이터에는 잘 작동하지 않을 수 있습니다. 이동 평균이 적절한 접근 방식인지 결정할 때는 데이터의 특정 특성과 탐지하려는 이상치의 성격을 고려해야 합니다.

## Moving Averages를 사용한 이상치 탐지 방법:

1. **Calculate Moving Average:**
   - 이동 평균은 지정된 창(window) 크기 내의 데이터 포인트들의 평균을 계산하여 얻습니다. 일반적으로 간단한 이동 평균(SMA), 가중 이동 평균(WMA), 지수 이동 평균(EMA) 등이 사용됩니다.

2. **Compare with Actual Data:**
   - 실제 데이터 포인트를 이동 평균과 비교합니다. 데이터 포인트가 이동 평균에서 크게 벗어날 경우, 해당 포인트를 이상치로 간주할 수 있습니다.

3. **Set Threshold:**
   - 이동 평균과의 차이를 기반으로 임계값을 설정합니다. 예를 들어, 이동 평균에서 특정 표준편차 범위를 벗어나는 포인트를 이상치로 정의할 수 있습니다.

4. **Detect Anomalies:**
   - 임계값을 초과하는 차이를 가진 데이터 포인트들을 이상치로 식별합니다.

## Moving Averages를 사용한 이상치 탐지의 장점:

1. **Noise Reduction:** 이동 평균은 시계열 데이터의 노이즈를 제거하여 기본 경향과 패턴을 식별하기 쉽게 만듭니다.
2. **Easy Implementation:** 이동 평균의 개념은 간단하고 구현이 쉬워서 다양한 수준의 기술을 가진 사용자들이 접근할 수 있습니다.
3. **Quick Identification of Trends:** 이동 평균은 데이터의 경향과 변화를 신속하게 식별하는 데 도움을 줍니다.
4. **Intuitive Interpretation:** 이상치는 이동 평균에서 크게 벗어난 데이터 포인트로 나타나며, 이는 직관적인 해석을 제공합니다.
5. **Robust to Short-Term Fluctuations:** 이동 평균은 단기 변동에 덜 민감하여 장기적인 이상치를 탐지하는 데 적합합니다.
6. **Versatility:** 이동 평균은 금융 데이터, 센서 판독값 등 다양한 시계열 데이터 유형에 적용될 수 있습니다.

## Moving Averages를 사용한 이상치 탐지의 단점:

1. **Lags Behind:** 이동 평균은 데이터 포인트 창을 기반으로 계산되므로 이상치 탐지에 지연이 발생할 수 있습니다. 이는 최근 이상치의 탐지를 지연시킬 수 있습니다.
2. **Difficulty with Rapid Changes:** 이동 평균은 데이터가 급격히 변화할 때 잘 반응하지 않을 수 있으며, 특히 창 크기가 큰 경우에 그러합니다.
3. **Inaccurate for Complex Patterns:** 복잡한 패턴을 가진 시계열의 경우, 이동 평균은 모든 변화를 효과적으로 포착하지 못하여 거짓 양성 또는 누락된 이상치를 초래할 수 있습니다.
4. **Dependent on Window Size:** 창 크기의 선택은 기법의 민감도에 영향을 미칩니다. 작은 창은 장기적인 이상치를 놓칠 수 있고, 큰 창은 단기적인 이상치를 놓칠 수 있습니다.
5. **Not Adaptive:** 이동 평균은 데이터 패턴의 변화에 잘 적응하지 않습니다. 데이터 패턴이 갑자기 바뀌면 이동 평균이 조정되는 데 시간이 걸릴 수 있습니다.
6. **Limited for Non-Linear Data:** 이동 평균은 선형성을 가정하므로, 비선형 경향을 가진 시계열에는 적합하지 않을 수 있습니다.
7. **Baseline Consideration:** 적절한 기준선 이동 평균(예: Simple, Weighted, Exponential)을 결정하는 것은 이상치 탐지에 영향을 미칠 수 있습니다.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset 생성
np.random.seed(42)
df = pd.DataFrame({
    'Feature1': np.random.normal(0, 1, 1000),
    'Feature2': np.random.normal(0, 1, 1000)
})

# 이상치를 데이터셋의 끝에 추가
df.iloc[-1] = [10, 10]
df.iloc[-2] = [-5, -5]

# 이동 평균 계산 (윈도우 크기: 50)
window_size = 50
moving_averages = df.rolling(window=window_size).mean()

# 이동 평균의 표준 편차 계산
moving_std = df.rolling(window=window_size).std()

# Z-점수 계산
z_scores = (df - moving_averages) / moving_std
# shape(z_scores)  # (1000, 2)

# 이상치 탐지를 위한 임계값 설정 (필요에 따라 조정)
threshold = 2.5

# 이상치 식별
anomalies = np.where(np.abs(z_scores) > threshold)
# anomalies.shape()  # (2,)
# type(anomalies)  # <class 'numpy.ndarray'>
print(anomalies)

# 데이터셋과 이상치 플로팅
plt.scatter(df['Feature1'], df['Feature2'], label='Normal')
plt.scatter(df.iloc[anomalies[0]]['Feature1'], df.iloc[anomalies[0]]['Feature2'], color='red', label='Anomalies')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.title('Moving Averages Anomaly Detection')
plt.legend()
plt.show()
````

# Seasonal Decomposition (time series approach)

Seasonal Decomposition은 시계열 데이터를 트렌드, 계절성, 잔차와 같은 기본 구성 요소로 분해하여 이상치를 탐지하는 기술입니다.

요약하면, Seasonal Decomposition은 시계열 데이터를 구성 요소로 분해하여 이상치를 식별하는 강력한 기술입니다. 그러나 그 효과는 분해의 품질과 주어진 데이터에 대한 분해 가정의 적합성에 달려 있습니다. 실제 데이터셋에 이 기술을 적용할 때는 장단점을 신중히 고려해야 합니다.

## Seasonal Decomposition 이상치 탐지 방법:

1. **Decompose Time Series:**
   - 시계열 데이터를 트렌드(Trend), 계절성(Seasonality), 잔차(Residual) 구성 요소로 분해합니다.
   - 일반적으로 `statsmodels`의 `seasonal_decompose` 함수를 사용합니다.

2. **Analyze Components:**
   - 분해된 구성 요소를 분석하여 트렌드와 계절성을 파악합니다.
   - 잔차는 트렌드와 계절성으로 설명되지 않는 부분을 나타냅니다.

3. **Identify Anomalies:**
   - 잔차를 분석하여 이상치를 식별합니다. 잔차가 크게 벗어난 포인트를 이상치로 간주할 수 있습니다.

## Seasonal Decomposition 장점:

1. **Pattern Separation:** Seasonal Decomposition은 시계열 데이터를 다양한 구성 요소로 분리하여 각각의 구성 요소와 전체 잔차에서 이상치를 식별하기 쉽게 만듭니다.
2. **Effective for Multiple Patterns:** Seasonal Decomposition은 트렌드, 계절성 및 불규칙 변동을 포함한 다양한 패턴을 캡처할 수 있습니다.
3. **Interpretability:** 분해된 구성 요소(트렌드, 계절성, 잔차)는 의미 있는 해석을 제공하여 이상치의 성격을 이해하는 데 도움을 줍니다.
4. **Highlighting Anomalies:** 트렌드와 계절성으로 설명되지 않는 잔차 구성 요소는 이러한 구성 요소에 의해 설명되지 않는 이상치를 효과적으로 강조할 수 있습니다.
5. **Flexible:** Seasonal Decomposition은 다양한 수준의 계절성과 트렌드를 가진 시계열 데이터에 잘 작동할 수 있습니다.
6. **Signal Enhancement:** 계절성과 트렌드를 제거함으로써, Seasonal Decomposition은 이러한 구성 요소에 의해 가려진 이상치와 패턴을 더 잘 식별하는 데 도움을 줍니다.

## Seasonal Decomposition 단점:

1. **Limited to Decomposition Quality:** Seasonal Decomposition의 효과는 분해 자체의 품질에 달려 있습니다. 부정확한 분해는 거짓 양성 또는 누락된 이상치를 초래할 수 있습니다.
2. **Assumption of Components:** Seasonal Decomposition은 시계열이 트렌드, 계절성, 잔차 구성 요소의 합으로 표현될 수 있다고 가정합니다. 이 가정은 모든 유형의 데이터에 적용되지 않을 수 있습니다.
3. **Parameter Tuning:** 계절성 주기와 같은 파라미터는 도메인 지식이나 시행착오를 통해 신중히 선택해야 하며, 이는 시간 소모적일 수 있습니다.
4. **Difficulty with Complex Patterns:** 기본 구성 요소(트렌드와 계절성)로 잘 캡처되지 않는 복잡한 이상치는 효과적으로 탐지되지 않을 수 있습니다.
5. **Inconsistent Performance:** Seasonal Decomposition은 명확한 패턴이 있는 일부 데이터셋에서는 잘 작동할 수 있지만, 노이즈가 많거나 불규칙한 데이터에서는 효과가 떨어질 수 있습니다.
6. **Lack of Adaptability:** Seasonal Decomposition은 데이터 패턴의 변화나 기본 구성 요소의 급격한 변화에 잘 적응하지 못할 수 있습니다.
7. **Data Length Requirement:** 일부 분해 방법은 구성 요소를 정확하게 추정하기 위해 최소 데이터 길이가 필요하여 짧은 시계열에는 적용이 제한될 수 있습니다.
8. **Subjectivity in Anomaly Definition:** 잔차 구성 요소에서 무엇이 이상치인지 결정하는 것은 주관적일 수 있으며, 응용 프로그램에 따라 다를 수 있습니다.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Sample dataset 생성
np.random.seed(42)
df = pd.DataFrame({
    'Feature1': np.random.normal(0, 1, 1000),
    'Feature2': np.random.normal(0, 1, 1000)
})

# 이상치를 데이터셋 중간에 추가
df.iloc[400] = [10, 10]
df.iloc[401] = [-5, -5]

# Seasonal Decomposition을 수행하는 함수 정의
def seasonal_decompose_anomaly_detection(data, freq=10):  # 주기 10
    decomposed = seasonal_decompose(data, period=freq)  # 데이터를 트렌드, 계절성, 잔차로 분해
    residual = data - decomposed.seasonal - decomposed.trend  # 잔차 계산 ( 원본 - 계절성, 트렌드 )
    return residual

# 각 특성에 Seasonal Decomposition 적용
df_decomposed = df.apply(seasonal_decompose_anomaly_detection, args=(10,))

# 이상치 탐지를 위해 잔차의 절대값(abs) 계산
df_residuals = df_decomposed.apply(np.abs)

# 이상치 탐지를 위한 임계값 정의
threshold = 2.5  # 데이터와 요구사항에 따라 이 임계값을 조정

# 이상치 식별
anomalies = np.where(df_residuals > threshold)

# 데이터셋과 이상치 플로팅
plt.scatter(df['Feature1'], df['Feature2'], label='Normal')  # 정상 데이터 포인트 플로팅
plt.scatter(df.iloc[anomalies[0]]['Feature1'], df.iloc[anomalies[0]]['Feature2'], color='red', label='Anomalies')  # 이상치 데이터 포인트 플로팅
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('Seasonal Decomposition Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

# Exponential Smoothing - (Time Series approach)

Exponential Smoothing(지수 평활법)은 시계열 예측 기법으로, 특히 데이터 패턴의 갑작스러운 변화와 관련된 이상치 탐지에 사용할 수 있습니다. Exponential Smoothing은 최근 데이터 포인트에 더 많은 가중치를 부여하여 시계열 데이터의 기본 트렌드와 패턴을 캡처하는 것을 목표로 합니다.

Exponential Smoothing은 시계열 데이터에서 예상 패턴으로부터의 갑작스러운 편차로 나타나는 이상치를 탐지하는 데 사용할 수 있습니다. 그 한계를 이해하고 분석되는 데이터의 특정 특성을 고려하는 것이 중요합니다.

## Exponential Smoothing을 사용한 이상치 탐지 방법:

1. **Basic Exponential Smoothing:**
   - Basic Exponential Smoothing (단일 지수 평활법)은 시계열의 노이즈를 제거하고 트렌드를 캡처하는 데 사용됩니다.
   - 과거 관찰값의 가중 평균을 계산하며, 시간이 지남에 따라 가중치는 감소합니다.

2. **Detecting Anomalies:**
   - 시계열 데이터의 이상치는 종종 예상 패턴에서의 갑작스러운 편차로 관찰될 수 있습니다. 이러한 편차는 오류, 이상치 또는 실제 이상 현상으로 인해 발생할 수 있습니다.

3. **Residual Calculation:**
   - 잔차는 관찰값과 Exponential Smoothing 모델이 예측한 값의 차이로 계산됩니다.
   - 큰 잔차는 모델의 예측이 실제 데이터와 크게 벗어나는 포인트를 나타냅니다.

4. **Thresholding:**
   - 잔차를 이상치로 간주할 임계값을 설정합니다.
   - 이 임계값을 초과하는 잔차는 이상 현상이나 비정상적인 이벤트를 나타낼 수 있습니다.

5. **Visualization:**
   - 시계열 데이터와 Exponential Smoothing 모델의 예측값을 플롯하고, 탐지된 이상치를 표시합니다.

## Exponential Smoothing을 사용한 이상치 탐지의 장점:

1. **Time Context:** Exponential Smoothing은 데이터의 시간적 맥락을 고려하므로, 역사적 패턴으로부터의 갑작스러운 편차로 발생하는 이상치를 탐지하는 데 적합합니다.
2. **Simplicity:** 이 방법은 이해하기 쉽고 구현하기 간단하여, 시계열 예측에 익숙한 사람들에게 적합합니다.
3. **Adaptive:** Exponential Smoothing은 시간 경과에 따른 변화하는 트렌드에 적응하므로, 변화하는 데이터 패턴에서 이상치를 탐지하는 데 유용합니다.

## Exponential Smoothing을 사용한 이상치 탐지의 단점:

1. **Limited to Time Series Data:** Exponential Smoothing은 시계열 데이터를 위해 설계되었으며, 다른 유형의 데이터에는 적합하지 않을 수 있습니다.
2. **Parameter Selection:** 적절한 평활 파라미터(예: 평활 계수, 계절성)를 선택하는 것이 정확한 이상치 탐지에 중요할 수 있습니다.
3. **Local Anomalies:** Exponential Smoothing은 트렌드나 계절성의 변화와 관련된 지역적 이상치를 탐지하는 데 가장 적합합니다. 글로벌 이상치 탐지에는 덜 효과적일 수 있습니다.
4. **Dependence on Data Quality:** Exponential Smoothing으로 탐지된 이상치는 기본 데이터의 품질과 일관성에 따라 달라집니다.


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Exponential Smoothing을 각 특성에 적용
smoothing_params = {'trend': None, 'seasonal': None, 'seasonal_periods': None}  # 평활 파라미터 설정
smoothed_data = {}  # 평활된 데이터를 저장할 딕셔너리

for column in df.columns:
    model = ExponentialSmoothing(df[column], **smoothing_params)  # Exponential Smoothing 모델 생성
    fit_model = model.fit()  # 모델 피팅
    smoothed_data[column] = fit_model.fittedvalues  # 피팅된 값을 저장

# 잔차 계산 (예측값과 실제값의 차이)
residuals = df - pd.DataFrame(smoothed_data)

# 이상치 탐지를 위한 임계값 설정 (필요에 따라 조정)
anomaly_threshold = 2.5

# 잔차를 기반으로 이상치 식별
anomalies = np.where(abs(residuals) > anomaly_threshold)

# 데이터셋과 이상치 플로팅
plt.scatter(df['Feature1'], df['Feature2'], label='Normal')  # 정상 데이터 포인트 플로팅
plt.scatter(df.iloc[anomalies[0]]['Feature1'], df.iloc[anomalies[0]]['Feature2'], color='red', label='Anomalies')  # 이상치 데이터 포인트 플로팅
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('Exponential Smoothing Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

# Histogram-Based Outlier Score (HBOS)

The Histogram-Based Outlier Score (HBOS)는 히스토그램을 사용하여 데이터 포인트의 확률 밀도를 근사하는 비지도 학습 이상치 탐지 알고리즘입니다. HBOS는 정상 데이터 포인트가 고밀도 영역에 발생하고 이상치는 저밀도 영역에 발생한다고 가정합니다. HBOS는 각 데이터 포인트에 대해 "outlier score"를 계산하며, 이는 해당 데이터 포인트의 밀도가 정상 데이터의 예상 밀도와 얼마나 다른지를 정량화한 것입니다.

HBOS는 데이터 포인트의 이상치 점수를 예상 분포로부터의 편차에 기반하여 추정하는 방법입니다. 각 특징 값의 히스토그램을 구성하고, 데이터 포인트가 속한 빈 높이의 합의 역수를 이상치 점수로 계산합니다. 이 점수는 데이터 포인트의 특징 값이 전형적인 분포와 얼마나 다른지를 반영하여, 이상치의 정도를 측정합니다.

HBOS는 고차원 데이터셋에서 특히 유용한 빠르고 효율적인 이상치 탐지 방법입니다. 구현이 용이하고 해석 가능한 결과를 제공합니다. 그러나 매개변수에 대한 민감성과 특징 독립성 가정은 실제 데이터에 적용할 때 고려해야 할 사항입니다. 탐색적 분석의 출발점이나 기본 방법으로는 좋지만, 그 효과는 데이터의 특성과 이상치의 성격에 따라 달라질 수 있습니다.

## HBOS의 장점:

1. **Simplicity:** HBOS는 이해하기 쉽고 구현하기도 간단합니다. 복잡한 수학적 연산이나 매개변수가 필요하지 않습니다.
2. **Efficiency:** 계산 효율성이 높아 대규모 데이터셋과 고차원 데이터에서 적합합니다.
3. **Scalability:** 고차원 공간에서도 잘 작동하여 많은 특징을 가진 애플리케이션에 적합합니다.
4. **Interpretability:** 이상치 점수는 이상치의 정도를 직접 반영합니다. 높은 점수는 이상치일 가능성이 높음을 나타냅니다.
5. **Unsupervised:** HBOS는 학습을 위해 라벨된 데이터가 필요하지 않습니다. 데이터의 분포를 활용하여 이상치를 탐지합니다.
6. **Handles Different Scales:** HBOS는 각 특징에 대해 독립적으로 히스토그램을 생성하므로 다양한 스케일의 특징을 잘 처리합니다.

## HBOS의 단점:

1. **Sensitivity to Parameters:** HBOS는 비교적 간단하지만, 히스토그램의 빈 수와 같은 매개변수 선택에 따라 성능이 민감하게 변할 수 있습니다.
2. **Assumes Independence:** HBOS는 특징의 독립성을 가정합니다. 이는 모든 경우에 성립하지 않을 수 있으며, 상관된 특징은 이상치 탐지 정확성에 영향을 미칠 수 있습니다.
3. **Limited to Numeric Data:** HBOS는 수치 데이터에서 가장 잘 작동합니다. 범주형 데이터나 혼합형 데이터에는 전처리가 필요할 수 있습니다.
4. **Limited for Complex Patterns:** HBOS는 단순한 히스토그램으로 잘 포착되지 않는 복잡한 이상치 패턴을 탐지하는 데 어려움을 겪을 수 있습니다.
5. **No Contextual Information:** HBOS는 데이터 포인트 간의 관계를 고려하지 않습니다. 히스토그램에서 이상치로 나타나지 않는 이상치는 놓칠 수 있습니다.
6. **No Adaptive Learning:** HBOS는 데이터 분포의 변화에 잘 적응하지 않습니다. 데이터 동적이 변화하면 재조정이 필요할 수 있습니다.
7. **Sensitive to Binning:** HBOS의 정확성은 빈 선택 전략에 영향을 받을 수 있습니다. 적절하지 않은 빈 크기를 선택하면 거짓 양성 또는 거짓 음성을 초래할 수 있습니다.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyod.models.hbos import HBOS

# 샘플 데이터셋 생성
np.random.seed(42)
df = pd.DataFrame({
    'Feature1': np.random.normal(0, 1, 1000),
    'Feature2': np.random.normal(0, 1, 1000)
})

# 데이터셋 중간에 이상치 추가
df.iloc[0] = [10, 10]  # 첫 번째 데이터 포인트를 이상치로 설정
df.iloc[1] = [-5, -5]  # 두 번째 데이터 포인트를 이상치로 설정

# HBOS 모델 생성
hbos = HBOS(contamination=0.1)  # 이상치 비율을 10%로 설정

# 모델 피팅
hbos.fit(df)

# 이상치 점수 예측
outlier_scores = hbos.decision_function(df)

# 이상치 식별
threshold = np.percentile(outlier_scores, 100 * (1 - 0.1))  # 90번째 백분위수로 임계값 설정
anomalies = df[outlier_scores > threshold]  # 이상치 점수가 임계값을 초과하는 데이터 포인트 식별

# 데이터셋과 이상치 플로팅
plt.scatter(df['Feature1'], df['Feature2'], label='Normal')  # 정상 데이터 포인트 플로팅
plt.scatter(anomalies['Feature1'], anomalies['Feature2'], color='red', label='Anomalies')  # 이상치 데이터 포인트 플로팅
plt.xlabel('Feature1')  # x축 라벨 추가
plt.ylabel('Feature2')  # y축 라벨 추가
plt.title('HBOS Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````

# PCA (Principal Component Analysis) for Anomaly Detection

PCA (Principal Component Analysis)는 이상치 탐지에 사용할 수 있지만, 그 효과는 데이터의 특성과 탐지하려는 이상치 유형에 따라 달라집니다. 다음은 PCA를 사용하여 이상치를 탐지할 때 고려해야 할 요소들입니다:

## PCA를 사용한 이상치 탐지의 장점:

1. **Dimensionality Reduction:** PCA는 데이터를 축소하면서 대부분의 분산을 유지하는 데 도움을 줍니다. 이는 복잡한 데이터를 단순화하고 중요한 특징을 강조하는 데 유용합니다.
2. **Noise Reduction:** PCA는 가장 중요한 변동을 포착하는 주성분에 집중함으로써 데이터의 노이즈 영향을 줄이는 데 도움을 줍니다.
3. **Visual Inspection:** 데이터를 2차원 또는 3차원(주성분)으로 축소할 때, 스캐터 플롯에서 데이터와 이상치를 시각화할 수 있어 통찰력을 제공할 수 있습니다.

## PCA를 사용한 이상치 탐지의 단점:

1. **Linear Transformation:** PCA는 선형 변환 기법입니다. 이상치가 주성분과 선형 관계를 나타내지 않으면 PCA는 이를 효과적으로 구분하지 못할 수 있습니다.
2. **Global Anomalies:** PCA는 정상 데이터의 중심에서 멀리 떨어진 전역 이상치를 탐지하는 데 적합하지 않을 수 있습니다. 대부분의 데이터 내 변동을 포착하는 데 중점을 둡니다.
3. **Distribution Assumptions:** PCA는 데이터가 Gaussian 분포를 따른다고 가정합니다. 이상치는 분포를 왜곡하고 PCA의 품질에 영향을 미칠 수 있습니다.
4. **Threshold Selection:** 원본 데이터와 재구성된 데이터 간의 잔차 오류(거리)를 기반으로 이상치를 탐지하는 임계값을 정의하는 것은 주관적이고 어려울 수 있습니다.
5. **High Dimensionality Requirement:** PCA는 고차원 데이터에서 더 효과적입니다. 특징이 몇 개 안 되는 경우 다른 방법이 더 효과적일 수 있습니다.

## PCA가 이상치 탐지에 효과적인 경우:

PCA는 주성분과 선형 관계를 나타내는 국부적 이상치에 더 효과적입니다. 정상 데이터의 분포에서 작은 편차를 나타내고, PCA로 포착된 기본 구조와 관련된 이상치 탐지에 유용할 수 있습니다. 고차원 데이터를 다룰 때 전처리 단계로 자주 사용됩니다.

비선형 관계를 가진 이상치나 정상 데이터와 크게 다른 이상치의 경우, Isolation Forest, One-Class SVM 또는 Autoencoder와 같은 다른 기법이 더 적합할 수 있습니다.

요약하자면, PCA는 이상치 탐지에 사용할 수 있지만, 데이터의 특성과 탐지하려는 이상치 유형을 고려하는 것이 중요합니다. PCA는 일부 경우에 잘 작동할 수 있지만, 모든 이상치 탐지 시나리오에 가장 적합한 선택은 아닐 수 있습니다.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 샘플 데이터셋 생성
np.random.seed(42)
df = pd.DataFrame({
    'Feature1': np.random.normal(0, 1, 1000),
    'Feature2': np.random.normal(0, 1, 1000),
    'Feature3': np.random.normal(0, 1, 1000),
    'Feature4': np.random.normal(0, 1, 1000)
})

# 데이터셋에 이상치 추가
df.loc[0] = [10, 12, 20, 13]  # 첫 번째 데이터 포인트를 이상치로 설정
df.loc[1] = [-5, -12, -15, -55]  # 두 번째 데이터 포인트를 이상치로 설정

# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# PCA 적용
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 주성분을 사용하여 데이터 재구성
X_reconstructed = pca.inverse_transform(X_pca)

# df.shape  # (1000, 4)
# X_reconstructed.shape  # (1000, 2)


# 잔차 계산 (원본 데이터와 재구성된 데이터 간의 차이)
residuals = np.sum(np.square(X_scaled - X_reconstructed), axis=1)

# 이상치 탐지를 위한 임계값 설정
threshold = 7  # 데이터와 원하는 민감도에 따라 이 값을 조정
anomalies = np.where(residuals > threshold)[0]

# 원본 데이터 포인트 플로팅
plt.scatter(df['Feature1'][2:], df['Feature2'][2:], label='Normal', color='blue')  # 정상 데이터 포인트 플로팅 (이상치 제외)
plt.scatter(df.loc[anomalies]['Feature1'], df.loc[anomalies]['Feature2'], label='Anomaly', color='red')  # 이상치 데이터 포인트 플로팅

plt.xlabel('Feature 1')  # x축 라벨 추가
plt.ylabel('Feature 2')  # y축 라벨 추가
plt.title('PCA Anomaly Detection')  # 그래프 제목 추가
plt.legend()  # 범례 추가
plt.show()  # 그래프 표시
````
